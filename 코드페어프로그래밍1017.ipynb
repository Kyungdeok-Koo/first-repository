```python
import random

class Account:
    bank_name = "SC은행"
    account_count = 0  # 생성된 계좌의 개수를 저장하는 클래스 변수

    def __init__(self, owner, balance):
        self.owner = owner  # 예금주
        self.balance = balance  # 초기 잔액
        self.account_number = self.generate_account_number()  # 계좌번호 생성
        self.deposit_count = 0  # 입금 횟수
        self.deposit_log = []  # 입금 내역
        self.withdraw_log = []  # 출금 내역

        # 계좌 생성 시마다 account_count를 1 증가
        Account.account_count += 1

    # 계좌번호를 3자리-2자리-6자리 형태로 랜덤 생성하는 메서드
    def generate_account_number(self):
        num1 = str(random.randint(100, 999))  # 3자리
        num2 = str(random.randint(10, 99))    # 2자리
        num3 = str(random.randint(100000, 999999))  # 6자리
        return f"{num1}-{num2}-{num3}"

    # 클래스 메서드: 생성된 계좌 수를 출력하는 메서드
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # 입금 메서드: 최소 1원 이상 입금 가능
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_log.append(amount)  # 입금 내역 기록
            print(f"{amount}원이 입금되었습니다. 현재 잔액: {self.balance:,}원")

            # 입금 횟수가 5회일 때 잔고의 1% 이자 지급
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += interest
                self.deposit_log.append(interest)  # 이자 지급 내역 기록
                print(f"이자 {interest:,}원이 지급되었습니다. 현재 잔액: {self.balance:,}원")
        else:
            print("입금은 최소 1원 이상 가능합니다.")

    # 출금 메서드: 잔액보다 많은 금액은 출금 불가
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족하여 출금할 수 없습니다.")
        else:
            self.balance -= amount
            self.withdraw_log.append(amount)  # 출금 내역 기록
            print(f"{amount}원이 출금되었습니다. 현재 잔액: {self.balance:,}원")

    # Q6: 계좌 정보를 출력하는 메서드
    def display_info(self):
        print(f"은행이름: {Account.bank_name}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.balance:,}원")

    # Q10: 입금 내역 출력 메서드
    def deposit_history(self):
        print(f"{self.owner}님의 입금 내역: {', '.join(str(amount) for amount in self.deposit_log)}")

    # Q10: 출금 내역 출력 메서드
    def withdraw_history(self):
        print(f"{self.owner}님의 출금 내역: {', '.join(str(amount) for amount in self.withdraw_log)}")


# Q8: 여러 객체 생성 및 리스트에 저장
accounts = [
    Account("홍길동", 500000),
    Account("이순신", 1500000),
    Account("강감찬", 2000000),
    Account("유관순", 800000)
]

# Q6: 각 계좌의 정보 출력
for account in accounts:
    account.display_info()

# Q9: 잔고가 100만원 이상인 고객 정보 출력
print("\n잔고가 100만원 이상인 고객 정보:")
for account in accounts:
    if account.balance >= 1000000:
        account.display_info()

# Q7: 입금 횟수 5회 시 이자 지급 테스트
accounts[0].deposit(500000)  # 입금 1회
accounts[0].deposit(200000)  # 입금 2회
accounts[0].deposit(300000)  # 입금 3회
accounts[0].deposit(400000)  # 입금 4회
accounts[0].deposit(100000)  # 입금 5회 (이자 지급)

# Q10: 입출금 내역 확인
accounts[0].withdraw(100000)  # 출금 1회
accounts[0].deposit_history()  # 입금 내역 확인
accounts[0].withdraw_history()  # 출금 내역 확인
```

```python
import random


class Account:
    bank_name = "SC은행"
    account_count = 0

    def __init__(self, owner, balance):
        self.owner = owner  # 예금주
        self.balance = balance  # 초기 잔액
        self.account_number = self.generate_account_number()  # 계좌번호 생성
        self.deposit_count = 0  # 입금 횟수
        self.deposit_log = []  # 입금 내역
        self.withdraw_log = []  # 출금 내역
        Account.account_count += 1

    # 계좌번호를 3자리-2자리-6자리 형태로 랜덤 생성하는 메서드
    def generate_account_number(self):
        num1 = random.randint(100, 999)  # 3자리
        num2 = random.randint(10, 99)    # 2자리
        num3 = random.randint(100000, 999999)  # 6자리
        return f"{num1}-{num2}-{num3}"

    # 계좌 정보를 출력하는 메서드
    def get_account_num(self):
        return self.account_number  # 계좌 번호 반환

    # 입금 메서드
    def deposit(self, money):
        if money >= 1:  # 입금은 최소 1원 이상 가능
            self.balance += money
            self.deposit_count += 1  # 입금 횟수 증가
            self.deposit_log.append(money)  # 입금 내역 기록
            print(f"{money:,}원이 입금되었습니다. 현재 잔액은 {self.balance:,}원 입니다.")

            # 입금 횟수가 5회일 때 잔고의 1% 이자 지급
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += interest
                self.deposit_log.append(interest)  # 이자 지급 내역 기록
                print(f'{interest:,}원이 지급되었습니다. 현재 잔액은 {self.balance:,}원 입니다.')
        else:
            print('입금은 최소 1원이상 가능합니다.')

    # 출금 메서드
    def withdraw(self, money):
        if money > self.balance:
            print("잔액 부족으로 출금할 수 없습니다.")
        else:
            self.balance -= money
            self.withdraw_log.append(money)  # 출금 내역 기록
            print(f"{money:,}원이 출금되었습니다. 현재 잔액은 {self.balance:,}원 입니다.")

    # 계좌 정보를 출력하는 메서드
    def display_info(self):
        print(f"은행이름: {Account.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

    # 입금 내역 출력 메서드
    def deposit_history(self):
        print(f'{self.owner}님의 입금 내역입니다.')
        if not self.deposit_log:
            print("입금 내역이 없습니다.")
        else:
            for amount in self.deposit_log:
                print(f"{amount:,}원")

    # 출금 내역 출력 메서드
    def withdraw_history(self):
        print(f"{self.owner}님의 출금 내역입니다.")
        if not self.withdraw_log:
            print("출금 내역이 없습니다.")
        else:
            for amount in self.withdraw_log:
                print(f"{amount:,}원")

```
